##### Dibujar mapa

```python
def dibujar_linea(imagen, punto_inicio, punto_fin, inicio):
    color = "green" 
    grosor = 2
    radio = 3
    draw = ImageDraw.Draw(imagen)
    draw.line([punto_inicio, punto_fin], fill=color, width=grosor)
    if inicio ==1:
        draw.ellipse(
            (punto_inicio[0] - radio, punto_inicio[1] - radio, punto_inicio[0] + radio, punto_inicio[1] + radio), fill="red")
```

### A

##### Ciudad mas cercana

```python
def heuristica(ciudad, df):
    if ciudad in df.index:
        # Utiliza 'idxmin' para encontrar la ciudad con la menor distancia a 'ciudad1'
        ciudad_menor_distancia = df.loc[ciudad].idxmin()
        distancia_minima = df.loc[ciudad, ciudad_menor_distancia]
        return ciudad_menor_distancia, distancia_minima
    else:
        print(f"{ciudad} no se encuentra en el DataFrame.")


heuristica("Santa Fe", df_base)
```

##### Heuristica

```python
imagen = Image.open("mapa.png")
df_ciudades = df_base
km_recorridos = 0
ciudad_inicial = "Santa Fe"
ciudad_i = ciudad_inicial
cantidad_de_columnas = df_base.shape[1]
recorridos = []
for i in range(1, cantidad_de_columnas):
    ciudad_proxima, distancia = heuristica(ciudad_i, df_ciudades)
    df_ciudades = df_ciudades.drop(ciudad_i, axis=1)
    km_recorridos += distancia
    data = [ciudad_i, ciudad_proxima, distancia, km_recorridos]
    recorridos.append(data)
    dibujar_linea(imagen, coordenadas[ciudad_i], coordenadas[ciudad_proxima], i)
    ciudad_i = ciudad_proxima

km_recorridos += df_base[ciudad_i][ciudad_inicial]
data = [ciudad_i, ciudad_inicial,df_base[ciudad_i][ciudad_inicial], km_recorridos]
recorridos.append(data)
dibujar_linea(imagen, coordenadas[ciudad_i], coordenadas[ciudad_inicial], i)

df_recorridos = pd.DataFrame(recorridos, columns=['Ciudad Inicial', 'Ciudad Final', 'Distancia (KM)', 'Total(KM)'])
imagen.save("imagen_con_lineas.png")
df_recorridos
```

```python
df_ciudades = df_base.copy()
cantidad_de_columnas = df_base.shape[1]
lista_nombres = list(df_base.head(0))
ciudades_recorridos = []	
for i in range(0, len(lista_nombres)):
    df_ciudades = df_base.copy()
    recorridos = []
    km_recorridos = 0
    ciudad_inicial = lista_nombres[i]
    ciudad_i = ciudad_inicial
    for j in range(1, cantidad_de_columnas):
        ciudad_proxima, distancia = heuristica(ciudad_i, df_ciudades)
        df_ciudades = df_ciudades.drop(ciudad_i, axis=1)
        km_recorridos += distancia
        data = [ciudad_i, ciudad_proxima, distancia, km_recorridos]
        recorridos.append(data)
        ciudad_i = ciudad_proxima

    km_recorridos += df_base[ciudad_i][ciudad_inicial] 
    data = [ciudad_i, ciudad_inicial, df_base[ciudad_i][ciudad_inicial], km_recorridos]
    recorridos.append(data)

    ciudad_kilometros_recorrido = [lista_nombres[i], km_recorridos, recorridos]
    ciudades_recorridos.append(ciudad_kilometros_recorrido)

lista_ordenada = sorted(ciudades_recorridos, key=lambda x: x[1])
df_recorridos = pd.DataFrame(lista_ordenada[0][2], columns=[
                             'Ciudad Inicial', 'Ciudad Final', 'Distancia (KM)', 'Total(KM)'])
df_recorridos
```
