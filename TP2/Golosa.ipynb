```python
import pandas as pd
```

```python

volumen_maximo = 4200 # cm3
peso_maximo = 3000 # gramos
mochila = []
# Lista de objetos

objetos = [
    # cm3, $
    [150,20],
    [325,40],
    [600,50],
    [805,36],
    [430,25],
    [1200,64],
    [770,54],
    [60,18],
    [930,46],
    [353,28],
]

objetos_por_peso = [
    # grs, $
    [1800, 72],
    [600, 36],
    [1200, 60],
]
```

```python
class Mochila(object):
    def __init__(self) -> None:
        self.volumen_actual = 0
        self.peso_actual = 0 
        self.objetos_guardados = []


    def agregarObjetoPorVolumen(self,objeto):
        """
        Agrega un objeto a la mochila.
        """
        global peso_maximo
        if volumen_maximo >= self.volumen_actual + objeto.volumen: 
            self.volumen_actual += objeto.volumen
            self.objetos_guardados.append(objeto)

    def agregarObjetoPorPeso(self,objeto):
        """
        Agrega un objeto a la mochila.
        """
        global peso_maximo
        if peso_maximo >= self.peso_actual + objeto.peso: 
            self.peso_actual += objeto.peso
            self.objetos_guardados.append(objeto)


    def calcular_precio_obtenido(self):
        """
        Muestra el precio total de los objetos guardados
        """
        precio = 0
        for objeto in self.objetos_guardados:
            precio += objeto.precio
        return precio

    def mostrar_objetos(self):
        """
        Muestra los objetos guardados
        """
        filas = []
        for objeto in self.objetos_guardados:
            filas.append([objeto.precio,objeto.volumen,objeto.peso,objeto.valor_calculado])
        return pd.DataFrame(filas,columns=["precio","volumen","peso","valor_calculado"])
```

```python
class Objeto(object):
    def __init__(self,precio,volumen=0,peso=0) -> None:
        self.volumen = volumen
        self.precio = precio
        self.peso = peso
        self.valor_calculado = self.calcular_valor()

    def calcular_valor(self):
        if self.volumen != 0:
            self.valor_calculado = self.precio/self.volumen
        if self.peso != 0:
            self.valor_calculado = self.precio/self.peso
        
        return self.valor_calculado  
```

# Ejercicio 2
Resolver el ejercicio anterior usando el algoritmo greedy y comentar su similitud o no con el exhaustivo.

```python
mochila = Mochila()
elementos = []
for objeto in objetos:
    volumen,precio = objeto
    nuevo_elemento = Objeto(volumen=volumen,precio=precio)
    elementos.append(nuevo_elemento)

elementos.sort(key=lambda x: x.valor_calculado, reverse=True)
precio = [] 

for objeto in elementos:
    mochila.agregarObjetoPorVolumen(objeto)
    
print("Volumen ocupado por los objetos: %s" % mochila.volumen_actual)
print("Precio obtenido: %s" % mochila.calcular_precio_obtenido())
mochila.mostrar_objetos() 
```
